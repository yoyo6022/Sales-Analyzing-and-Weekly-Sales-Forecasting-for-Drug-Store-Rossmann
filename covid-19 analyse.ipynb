## Predictor(use Total cases to predict total deaths)

```python
import pandas as pd
import numpy as np

df = pd.read_csv('owid-covid-data.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df[df.location != 'World']
```

```python
pdf = df.pivot_table(['total_cases', 'total_deaths'], index='date',aggfunc='sum')
pdf.tail(5)
```

```python
x = np.array(pdf['total_cases'].reset_index(drop=True))
```

```python
y = np.array(pdf['total_deaths'].reset_index(drop=True))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=5)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train= x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import Lasso
model = Lasso()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict_train = model.predict(x_train)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_train,y_predict_train))
```

```python
y_predict_test = model.predict(x_test)
```

```python
print(r2_score(y_test,y_predict_test))
```
